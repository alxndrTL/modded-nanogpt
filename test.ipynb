```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
num_iterations = 5078 #3052 #4578
slw_start = 8
slw_end = 1024
slw_iterations = 0

def get_seqlen(it):
    if it < slw_iterations:
        seqlen = slw_start + ((it)/slw_iterations) * (slw_end - slw_start)
    else:
        seqlen = slw_end
    seqlen = int(seqlen)
    return seqlen - seqlen%8
```

```python
seq_len = slw_start
seqs = []

for t in range(num_iterations):
    seqs.append(get_seqlen(t))
```
