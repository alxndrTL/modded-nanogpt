```python
num_iterations = 3728 #5078 #3052 #4578
slw_start = 8
slw_end = 1024
slw_iterations = 972

def get_seqlen(it):
    if it < slw_iterations:
        seqlen = slw_start + ((it)/slw_iterations) * (slw_end - slw_start)
    else:
        seqlen = slw_end
    seqlen = int(seqlen)
    return seqlen - seqlen%8

seq_len = slw_start
seqs = []

for t in range(num_iterations):
    seqs.append(get_seqlen(t))

sum([s*512 for s in seqs])
```
